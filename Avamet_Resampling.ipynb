```python
path_2021= '/content/drive/MyDrive/Colab Notebooks/Data/Avamet/Temperature/2021/Common Stations/'
save_path_res_21= '/content/drive/MyDrive/Colab Notebooks/Data/Avamet/Temperature/2021/Resampled'

for file in os.listdir(path_2021):
  df = pd.read_csv(os.path.join(path_2021, file))
  df['timestamp']=pd.to_datetime(df['timestamp'])
  df.index=df['timestamp']
  df=df.resample('10min').mean()
  df.to_csv(os.path.join(save_path_res_21, file))
```

```python
def check_nan_in_temp_value(folder_path):
    # List all files in the directory and sort them by name
    files = sorted([f for f in os.listdir(folder_path) if f.endswith('.csv')])

    # Initialize a list to store results
    results = []

    # Iterate over each file
    for file in files:
        file_path = os.path.join(folder_path, file)

        # Read the CSV file
        try:
            df = pd.read_csv(file_path)

            # Check if the temp_value column exists
            if 'temp_value' in df.columns:
                # Count NaN values in the temp_value column
                nan_count = df['temp_value'].isna().sum()
            else:
                nan_count = None  # Use None if the column doesn't exist

            # Append the result to the list
            results.append({'file_name': file, 'missing_values': nan_count})

        except Exception as e:
            # If there is an error reading the file, record it as well
            results.append({'file_name': file, 'missing_values': f'Error: {e}'})

    # Create a DataFrame from the results
    result_df = pd.DataFrame(results)

    return result_df

# Example usage:
folder_path = folder_path ="/content/drive/MyDrive/Colab Notebooks/Data/Avamet/Temperature/2021/Resampled"
result_df_21 = check_nan_in_temp_value(folder_path)
```

# Missing Values Thresholding

```python
def check_missing_values_threshold(folder_path, threshold=0.05):
    # List all files in the directory and sort them by name
    files = sorted([f for f in os.listdir(folder_path) if f.endswith('.csv')])

    results = []

    for file in files:
        file_path = os.path.join(folder_path, file)

        try:
            df = pd.read_csv(file_path)
            if 'timestamp' in df.columns and 'temp_value' in df.columns:
                total_values = df['timestamp'].count()

                missing_values = df['temp_value'].isna().sum()
                missing_percentage = missing_values / total_values
                exceeds_threshold = missing_percentage > threshold

                # Append the result to the list
                results.append({
                    'file_name': file,
                    'total_values': total_values,
                    'missing_values': missing_values,
                    'missing_percentage': missing_percentage,
                    'exceeds_threshold': exceeds_threshold
                })
            else:
                results.append({
                    'file_name': file,
                    'total_values': None,
                    'missing_values': None,
                    'missing_percentage': None,
                    'exceeds_threshold': 'Columns missing'
                })

        except Exception as e:
            # If there is an error reading the file, record it as well
            results.append({
                'file_name': file,
                'total_values': None,
                'missing_values': None,
                'missing_percentage': None,
                'exceeds_threshold': f'Error: {e}'
            })

    # Create a DataFrame from the results
    result_df = pd.DataFrame(results)

    return result_df

# Example usage:
folder_path = "/content/drive/MyDrive/Colab Notebooks/Data/Avamet/Temperature/2022/Resampled/"
result_df = check_missing_values_threshold(folder_path)

# # Display the DataFrame
# print(result_df)
```

```python
import os
import pandas as pd

# Function to process each file and count missing data
def process_file(file_path, file_name):
    df = pd.read_csv(file_path)

    # Convert the timestamp columns to datetime
    df['timestamp'] = pd.to_datetime(df['timestamp'])

    # Check for missing values in the temperature column
    df['is_missing'] = df['temp_value'].isnull()

    # Create a group to count consecutive missing values
    df['missing_group'] = (df['is_missing'] != df['is_missing'].shift()).cumsum()

    # Filter only the missing data groups
    missing_data_df = df[df['is_missing']]

    # Count missing values in each gap
    missing_counts = missing_data_df.groupby('missing_group')['temp_value'].size().reset_index(name='missing_count')

    # Add a column with the file name for identification
    missing_counts['file_name'] = file_name

    return missing_counts

# Function to process all files in a directory and save results in one file
def process_all_files_in_folder(folder_path, output_file):
    file_list = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

    all_missing_counts = []

    for file_name in file_list:
        file_path = os.path.join(folder_path, file_name)
        missing_counts = process_file(file_path, file_name)
        all_missing_counts.append(missing_counts)

    # Concatenate all the missing counts into a single DataFrame
    combined_missing_counts = pd.concat(all_missing_counts, ignore_index=True)

    # Save the combined missing counts to a single CSV file
    combined_missing_counts.to_excel(output_file, index=False)
    print(f"All missing counts saved to {output_file}")

# Specify the folder path where the files are located
folder_path = "/content/drive/MyDrive/Colab Notebooks/Data/Avamet/Temperature/2023/Resampled/"

# Specify the output file where the combined results will be saved
output_file = 'combined_missing_data_counts_23.xlsx'  # You can change the name if needed

# Process all files in the folder and save to one file
process_all_files_in_folder(folder_path, output_file)
```

```python
import os
import re

def list_csv_file_names(folder_path):
    # Get all files in the folder
    all_files = os.listdir(folder_path)

    # Filter for .csv files, remove the extension and `_2023` part
    csv_file_names = [
        re.sub(r"_\d{4}$", "", os.path.splitext(file)[0])  # Remove `_2023` or similar patterns
        for file in all_files if file.endswith('.csv')
    ]
    csv_count = len(csv_file_names)
    return csv_file_names, csv_count
# Example usage
folder_path =  "/content/drive/MyDrive/Colab Notebooks/Data/Avamet/Temperature/2023/Interpolated XGBoost"  # Replace with the path to your folder
csv_names, csv_count = list_csv_file_names(folder_path)
print("CSV file names without extensions:", csv_names)
print("Total number of CSV files:", csv_count)
```
